Then we create a `.txt` file named as `op.txt` in which we will copy the data from `.xls` into text format to perform operation the text.

Now we load the `.txt` file to perform the tokenization, removal of punchuations and filtering the numbers. We also filter the `char` to make our output more clean. Finally we the print the desired output and export in the output into a `.xls` file named `op.xls`.

```
filename = 'op.txt' 
file = open(filename, 'rt') # opening the text file in read mode.
Lines = file.readlines() # Number of line present in the text file (op.txt)
for line in Lines: 
  base_text = line.strip() # convertung list into string
  words = base_text.split() # breaking string into the words
  words = re.split(r'\W+', base_text) # splitting the words again to split the punctuations
  re_punc = re.compile('[%s]' % re.escape(string.punctuation)) # Compiling the punctuations into a single variable
  # remove punctuation from each word
  stripped = [re_punc.sub('', w) for w in words]
  re_print = re.compile('[^%s]' % re.escape(string.printable)) # removing the punctuations
  result = [re_print.sub('', w) for w in words] # storing the words into the a variable after removing the punctuations from the words
  sentences = sent_tokenize(base_text) # tokenizing the sentences
  tokens = word_tokenize(base_text) # tokenizing the words
  # remove all tokens that are not alphabetic
  words = [word for word in tokens if word.isalpha()]
  #split into words
  tokens = word_tokenize(base_text)
  # convert to lower case
  tokens = [w.lower() for w in tokens]
  # prepare regex for char filtering
  re_punc = re.compile('[%s]' % re.escape(string.punctuation))
  # remove punctuation from each word
  stripped = [re_punc.sub('', w) for w in tokens]
  # remove remaining tokens that are not alphabetic
  words = [word for word in stripped if word.isalpha()]
  # filter out stop words
  stop_words = set(stopwords.words('english')) 
  words = [w for w in words if not w in stop_words] # removal of stopwords
  print(base_text ,'\t\t -> \t\t', words[:100]) # Prints out the result in the output panel below
  for word in words: 
    excel = word.strip()
    excel = open('op.xls', 'w+')  # writing the text file into the excel sheet names op.xls
    excel.write(word)
file.close() # closing the file to save the file.
```
